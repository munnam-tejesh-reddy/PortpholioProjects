```python
# Importing the packages that are going to be used in project

import pandas as pd

import numpy as np
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (14,8)#justs the config of the plot we create.

pd.options.mode.chained_assignment = None
```

```python
# Reading the data from local c drive.

df = pd.read_csv(r'E:\Data Analysis\Portfolio Project\Project in python on movies data\movies.csv')
```

```python
# Loading data of first ten rows to view data.

df.head()
```

```python
# Finding missing data by creating a loop

for col in df.columns :
    perct_missing = np.mean(df[col].isnull())
    print ('{} - {}%'.format(col, perct_missing))
```

```python

# Filling the missing data using forward fill command
new_df = df.fillna(method='ffill')
```

```python
# Rechecking if still there is any missing data.

for col in new_df.columns :
    perct_missing = np.mean(new_df[col].isnull())
    print ('{} - {}%'.format(col, perct_missing))
```

```python
df.head()
```

```python
# Sorting the data by gross values in descending order.

new_df= new_df.sort_values(by= ['gross'],inplace= False, ascending= False)
```

```python
# Seperating 'Year' from the released column.

new_df[['released month','released year']]=df['released'].str.rsplit(',', n=2,expand= True)
new_df[['Myear','placed']]=new_df['released year'].str.split('(', n=2, expand= True)
```

```python
new_df.head()
           
```

```python
#droping unwanted/unneccessary columns.

new_df=new_df.drop('placed',axis= 1)
new_df=new_df.drop('released',axis= 1)
new_df=new_df.drop('released month',axis= 1)
new_df=new_df.drop('released year',axis= 1)
new_df=new_df.drop('year',axis= 1)
```

```python
new_df.head()
```

```python
new_df['gross']=new_df['gross'].astype('int64')
new_df['budget']=new_df['budget'].astype('int64')
```

```python
numerized_df= new_df
```

```python
for col_name in numerized_df.columns:
    if (numerized_df[col_name].dtype == 'object' and numerized_df[col_name].dtype !='float64'):
        numerized_df[col_name]=  numerized_df[col_name].astype('category')
        numerized_df[col_name]= numerized_df[col_name].cat.codes
```

```python
# Finding the correlatin  using UNSTACK to see corrleation for each column comparing to other columns itslef.

correlation_matrix = numerized_df.corr()

correlation_pair = correlation_matrix.unstack()
correlation_pair
```

```python
correlation_pair.sort_values()
```

```python
sorted_pair=correlation_pair.sort_values()

```
